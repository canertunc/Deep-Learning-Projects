**I created a model by analyzing bank data for artificial neural network  learning purposes.**

```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
X = data.drop(["CustomerId","Surname","Exited","RowNumber"],axis = 1)
```

```python
X.head()
```

```python
y = data["Exited"]
```

```python
y.value_counts()
```

```python
colm = ["Geography","IsActiveMember","HasCrCard","Gender"]
```

```python
X = pd.get_dummies(X,columns=colm)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=58)
```

```python
cols = X_train.columns
cols
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
import tensorflow as tf
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

print(accuracy_score(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```
